```python
'/scratch/az338/ucc-fileserver/AZ_challenge_data/drug_synergy_data/'
DATA_DIR = '/scratch/az338/ucc-fileserver/AZ_challenge_data/'
MOL_DATA_DIR = DATA_DIR + 'drug_synergy_data/'

data = pd.read_csv(MOL_DATA_DIR+'ch1_train_combination_and_monoTherapy.csv')
data.head()
```

```python
targets = data.COMPOUND_A.tolist()+data.COMPOUND_B.tolist()
targets=[sub('_([0-9]|[A-Z]*)','',x) for x in targets]
target_counts = Counter(targets)
target_counts = pd.DataFrame({'N_COMPOUND':pd.Series(target_counts.values(),index=target_counts.keys())})
target_counts.sort('N_COMPOUND').plot(kind='barh')
```

```python
challenge_targets = target_counts.index.tolist()
f_v1 = open('/scratch/az338/ucc-fileserver/AZ_challenge_data/pidgin_symbol_v1.txt','r')
pidgin_targets_v1 = [t.rstrip('\n') for t in f_v1]
f_v1.close()
f_v2 = open('/scratch/az338/ucc-fileserver/AZ_challenge_data/pidgin_symbol_v2.txt','r')
pidgin_targets_v2 = [t.rstrip('\n') for t in f_v2]
f_v2.close()


inter1 = set(challenge_targets).intersection(pidgin_targets_v1)
inter2 = set(challenge_targets).intersection(pidgin_targets_v2)
print len(inter1), len(inter2), len(challenge_targets)
```
